```python
#Histogram
#Plots in matplotlib reside within a figure object, use plt.figure to create new figure
fig=plt.figure() 

#Create one or more subplots using add_subplot, because you can't create blank figure
viz = fig.add_subplot(1,1,1)

#Variable
viz.hist(df['Age'],bins = 10) # Here you can play with number of bins - what is bin? A bucket.

#Labels and Title
plt.title('Age distribution by bin')
plt.xlabel('Age')
plt.ylabel('No of Employee in each bin')
plt.show()
fig.savefig("test4.png")
```

```python
#Boxplot - Understand about boxplot
fig=plt.figure()
viz = fig.add_subplot(1,1,1)
#Variable
viz.boxplot(df['Age'])
plt.title('Boxplot for Age')
plt.show()
fig.savefig("test3.png")
```

```python
#scatter plot
fig = plt.figure()
viz = fig.add_subplot(1,1,1)
viz.scatter(df['Age'],df['Sales']) #You can also add more variables here to represent color and size.
#Labels and Title
plt.title('Age vs Sales')
plt.xlabel('Age')
plt.ylabel('Sales')
plt.show()
fig.savefig("test2.png")
```

```python
#Bubble plot
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(df['Age'],df['Sales'], s=df['Income']) # Added third variable income as size of the bubble
plt.show()
fig.savefig("test1.png")
```

```python
#line plot
import numpy as np
# Creating data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure and then viz = fig.add_subplot(111)
fig, viz = plt.subplots()
viz.plot(t, s)

#Set axes level and title
viz.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='A simple plot')

#Create grid in chart
viz.grid()

#saving the plot in current working directory
fig.savefig("test.png")

#Showing the plot

plt.show()
```

```python
#barplot
A = [5., 30., 45., 22.]
B = [5., 25., 50., 20.]

X = range(4)

#Plain bar
plt.bar(X, A, color = 'b')

#stacked bar
plt.bar(X, B, color = 'r', bottom = A)
plt.show()
```

```python
#Pie Chart

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=0)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
fig1.savefig("testPie.png")
```

```python
#Create two plots together
import numpy as np
import matplotlib.pyplot as plt
#creating dummy data for x axis
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

#Creating dummy data for y axis
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

fig = plt.figure()
#Defining the subplot - 1
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('2 subplots')
plt.ylabel('Damp oscillation')

#Defining the subplot - 1
plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamp oscillation')

plt.show()
fig.savefig("testVilla.png")
```

```python
#Few nice plots using seaborn
import seaborn as sns

#sns.set(style="white", color_codes=True)

#load the Iris flower dataset
iris = pd.read_csv("Iris.csv") # the iris dataset is now a Pandas DataFrame

# Let's see what's in the iris data
iris.head()
```

```python
sns.distplot(iris["SepalWidth"])
```

```python
# Histogram by sns
sns.distplot(df['Age'], kde=False, bins=4)
```

```python
sns.distplot(iris["SepalWidth"], kde=False, bins=15)
```

```python
#SepalLength	SepalWidth	PetalLength	PetalWidth
sns.distplot(iris["SepalLength"], kde=False, bins=25)
```

```python
# Fitting a parametric Distribution!
import scipy.stats as stats
sns.distplot(iris["SepalLength"], kde=False, bins=25, fit=stats.gamma)
```

```python
iris["SepalWidth"].min()
```

```python
# A seaborn jointplot - It shows bivariate scatterplots and univariate histograms in the same figure
#SepalLength	SepalWidth	PetalLength	PetalWidth	Name
sns.jointplot(x="SepalLength", y="SepalWidth", data=iris, height=5)
plt.show()
```

```python
# One piece of information missing in the plots above is what species each plant is
# We'll use seaborn's FacetGrid to color the scatterplot by species
sns.FacetGrid(iris, hue="Name", size=5) \
   .map(plt.scatter, "SepalLength", "SepalWidth") \
   .add_legend()
plt.show()
```

```python
# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Name", y="PetalLength", data=iris)
plt.show()
```

```python
# One way we can extend this plot is adding a layer of individual points on top of
# it through Seaborn's striplot
# 
# We'll use jitter=True so that all the points don't fall in single vertical lines
# above the species
#
# Saving the resulting axes as ax each time causes the resulting plot to be shown
# on top of the previous axes
ax = sns.boxplot(x="Name", y="PetalLength", data=iris)
ax = sns.stripplot(x="Name", y="PetalLength", data=iris, jitter=True, edgecolor="gray")
plt.show()
```

```python
# A violin plot combines the benefits of the previous two plots and simplifies them
# Denser regions of the data are fatter, and sparser thiner in a violin plot
sns.violinplot(x="Name", y="PetalLength", data=iris, size=6)
plt.show()
```

```python
# A final seaborn plot useful for looking at univariate relations is the kdeplot,
# which creates and visualizes a kernel density estimate of the underlying feature
sns.FacetGrid(iris, hue="Name", size=6) \
   .map(sns.kdeplot, "PetalLength") \
   .add_legend()
plt.show()
```

```python
# Another useful seaborn plot is the pairplot, which shows the bivariate relation
# between each pair of features
sns.pairplot(iris, height=3)
plt.show()
```

```python
# KDE plot in the diagonal
sns.pairplot(iris, height=3, diag_kind="kde")
plt.show()
```

```python
# From the pairplot, we'll see that the Iris-setosa species is separataed 
# from the other two across all feature combinations
sns.pairplot(iris, hue="Name", size=3)
plt.show()
```

```python
# The diagonal elements in a pairplot show the histogram by default
# We can update these elements to show other things, such as a kde
sns.pairplot(iris, hue="Name", size=3, diag_kind="kde")
plt.show()
```
